```python
import sys
sys.path.insert(0, "..")
from classes.stat_phys.potentials1 import H_Metro_MonteCarlo as H_metro
import matplotlib.pyplot as plt
import numpy as np
plt.rc('animation', html='jshtml')
```

```python
def proposal_func(x, delta=0.1):
    return x + delta*np.random.randn(1)

def p_fun(x,T,k,x_0):
    t1 = np.sqrt(k/(2.0*np.pi*T))
    t2 = np.exp(-1.0/2.0*k*(x-x_0)**2/T)
    return t1*t2

xs = np.linspace(-4.0,4.0, 100)

def v_fun(x, k, x_0):
    return 1.0/2.0*k*(x-x_0)**2


points = [100, 1000, 10000, 100000]
H_pots = []
prop_args = {"delta":0.1}
for N in points:
    v_fun_args = {"k":1.0, "x_0":0.0}
    pot = H_metro(T=0.15, v_func=v_fun, v_func_args=v_fun_args, x0=0.0, proposal_func=proposal_func, proposal_func_args=prop_args, N=N)
    H_pots.append(pot)

fig, axs= plt.subplots(1, 4, figsize=(15,3), layout="constrained")
#Y = lambda x, T, k, x_0, x_min, x_max: np.sqrt(k/(2.0*np.pi))*np.exp(-k/2.0*(x-x_0)**2)/special.erf(x_max*np.sqrt(k/2.0)) #+ 1.0/(x_max-x_min)
for ax, pot, point in zip(axs, H_pots, points):
    ax.set_title(r"$N=$"+f"{point}")
    ax.grid()
    ax.set_xlabel(r"$x$")
    v_fun_args = {"k":1.0, "x_0":0.0}
    ax.set_ylabel(r"$V(x)$ $[k_BT_{0}]$")
    ax.set_xlim([-3.0,3.0])
    ax.set_ylim([0,3])
    pot.plot(ax, plot_range=[-3.0,3.0])
    ax.plot(xs, p_fun(xs, T=pot.T, **v_fun_args), c="C1", label=r"$P(x)-$Analytical")
    ax.legend()
```
