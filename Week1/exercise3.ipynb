```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np

plt.rc('animation', html='jshtml')
```

```python
class Trekant():
    def __init__(self, points, color="C0") -> None:
        self.color = color
        self.points = points.copy()
        self.cm = np.mean(self.points, axis=0)
        self.triangle = None
        self.cross = None

    def draw(self, ax):
        if self.triangle == None:
            self.triangle = ax.fill(self.points[:,0], self.points[:,1], color=self.color)[0]
        if self.cross == None:
            self.cross = ax.plot(self.cm[0], self.cm[1], "x", c='k', ms=20, mew=3)[0]
        else:
            self.move(np.array([0.0,0.0]))

    def move(self, r):
        for points in self.points:
            points+=r
        self.cm = np.mean(self.points, axis=0)
        self.cross.set_data(self.cm[0], self.cm[1])
        self.triangle.update({"xy":self.points, "color":self.color})
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.grid()
ax.set_xlim([0,8])
ax.set_ylim([0,5])
```

```python
P1 = np.array([2.0,1.0])
P2 = np.array([4.0,3.0])
P3 = np.array([1.0,4.0])
points = np.array([P1,P2,P3])
t = Trekant(points)
t.draw(ax)
fig
```

```python
r = np.array([2,1])
t.move(r)
fig
```
