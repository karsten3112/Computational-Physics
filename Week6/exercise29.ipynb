```python
import copy
import sys
sys.path.insert(0, "..")
from classes.regression.regressers import Lin_Regression, Pol_Regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.grid()
ax.set_xlim([-3,6])
ax.set_ylim([-3,10])
ax.axhline(0, c="k")
ax.axvline(0, c="k")
def lin_func(xs, a=1.2, b=2.0):
    return xs*a + b

xs = np.linspace(-2,5, 1000)
x_tab = np.array([-1, 1, 2,3 ,4.5])
data = np.zeros(shape=(len(x_tab), 2))
data[:,0] = x_tab
data[:,1] = lin_func(x_tab)
ax.plot(xs, lin_func(xs), c='r', label="underlying function from sampling")
ax.plot(data[:,0], data[:,1], 'o', c="r", label="sampled points")

lin_reg = Lin_Regression(data=data)
x_data_sample = np.random.rand(20)*7 - 2
y_data_pred = lin_reg.sample(x_data=x_data_sample)
ax.plot(x_data_sample, y_data_pred, 'o', c="g", ms=5, alpha=0.8, label="prediction (Linear regression)")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.grid()

ax.set_xlim([-3,3])
ax.set_ylim([-3,12])
ax.axhline(0, c="k")
ax.axvline(0, c="k")
def quad_func(xs, a=2.5, b=-1, c=-2.0):
    return a*xs**2 + b*xs + c

xs = np.linspace(-2,2, 1000)
x_tab = np.array([-1, 1, 2,-1.5, 0.0])
data = np.zeros(shape=(len(x_tab), 2))
data[:,0] = x_tab
data[:,1] = quad_func(x_tab)
ax.plot(xs, quad_func(xs), c='r', label="underlying function from sampling")
ax.plot(data[:,0], data[:,1], 'o', c="r", label="sampled points")

lin_reg = Lin_Regression(data=data)
pol_reg = Pol_Regression(data=data, N_deg=2)
x_data_sample = np.random.rand(20)*4 - 2
y_data_lin = lin_reg.sample(x_data=x_data_sample)
y_data_pol = pol_reg.sample(x_data=x_data_sample)
ax.plot(x_data_sample, y_data_lin, 'o', c="g", ms=5, alpha=0.8, label="prediction (Linear regression)")
ax.plot(x_data_sample, y_data_pol, 'o', c="blue", ms=5, alpha=0.8, label="prediction (pol. regression)")
ax.legend()
```
