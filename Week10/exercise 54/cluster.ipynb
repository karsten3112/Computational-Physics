```python
import torch
import sys
sys.path.insert(0, "../../")
import matplotlib.pyplot as plt
import numpy as np
from classes.nn_classes import train_model, NNmodel
from torch.utils.data import Dataset, DataLoader
from classes.atom_classes.atom_collection import Atom, Atom_Collection
from classes.atom_classes.atom_collection_plotter import Atom_Collection_Plotter
```

```python
data_flat = np.loadtxt('../pos_e1_e2_e3_1000.txt')
data_in_file = [[d[:24].reshape(12,2),d[24],d[25],d[26]] for d in data_flat]

ENERGY_LABEL = 1
```

```python
fig, axs = plt.subplots(1,4, figsize=(10,3), layout="constrained")
atom_plotter = Atom_Collection_Plotter(markersize=20)
for data, ax in zip(data_in_file, axs):
    ax.grid()
    ax.set_xlim([2.0, 8.0])
    ax.set_ylim([1.0, 8.0])
    ax.set_aspect("equal")
    positions = data[0]
    atom_col = Atom_Collection(atomlist=[Atom(position=pos) for pos in positions])
    atom_plotter(atom_col=atom_col, ax=ax)
    ax.set_title(r"$E=$"+f"{round(data[1],3)}")
```
